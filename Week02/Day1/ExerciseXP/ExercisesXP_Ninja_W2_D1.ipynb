## 3. Data Understanding

```python
# Description: Load dataset, check structure, data types, and missing values.

import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("=== Dataset Shape ===")
print(df.shape)

print("\n=== First 5 rows ===")
display(df.head())

print("\n=== Info ===")
df.info()

print("\n=== Missing values per column ===")
print(df.isna().sum())

print("\n=== Descriptive statistics (numeric) ===")
display(df.describe())

print("\n=== Descriptive statistics (categorical) ===")
display(df.describe(include=['object']))
```

## 4. Data Cleaning

```python
# Description: Defensive cleaning with explicit dtypes, safe imputations, and stable encoding.

import pandas as pd
import numpy as np

# Optional: if scikit-learn is missing in your env, uncomment the next line
# !pip install scikit-learn -q

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# --- Copy and basic type coercion ---
df_clean = df.copy(deep=True)

# Ensure expected columns exist
expected_cols = ["Age", "Embarked", "Cabin", "Sex", "Fare"]
missing_expected = [c for c in expected_cols if c not in df_clean.columns]
if missing_expected:
    raise KeyError(f"Missing expected columns: {missing_expected}")

# Coerce numeric columns safely (non-numeric -> NaN)
for num_col in ["Age", "Fare", "SibSp", "Parch", "Pclass", "Survived"]:
    if num_col in df_clean.columns:
        df_clean[num_col] = pd.to_numeric(df_clean[num_col], errors="coerce")

# Normalize text columns to string (avoids unexpected types from CSVs)
for txt_col in ["Sex", "Embarked", "Cabin", "Ticket", "Name"]:
    if txt_col in df_clean.columns:
        df_clean[txt_col] = df_clean[txt_col].astype("string")

# --- Imputations ---
# Age: median
if "Age" in df_clean.columns:
    age_imp = SimpleImputer(strategy="median")
    df_clean["Age"] = age_imp.fit_transform(df_clean[["Age"]]).ravel()

# Fare: median (in case source CSV has missing Fare)
if "Fare" in df_clean.columns:
    fare_imp = SimpleImputer(strategy="median")
    df_clean["Fare"] = fare_imp.fit_transform(df_clean[["Fare"]]).ravel()

# --- Embarked: most frequent (mode) ---
if "Embarked" in df_clean.columns:
    # Convert to object dtype and replace pandas <NA> with np.nan
    df_clean["Embarked"] = df_clean["Embarked"].astype("object").replace({pd.NA: np.nan})

    if df_clean["Embarked"].isna().all():
        # If all values are missing, fill with default 'S'
        df_clean["Embarked"] = "S"
    else:
        emb_imp = SimpleImputer(strategy="most_frequent")
        df_clean["Embarked"] = emb_imp.fit_transform(df_clean[["Embarked"]]).ravel()

    # Ensure final string type
    df_clean["Embarked"] = df_clean["Embarked"].astype(str)

# --- Feature indicators ---
# HasCabin: 1 if any cabin string present, else 0
if "Cabin" in df_clean.columns:
    df_clean["HasCabin"] = df_clean["Cabin"].notna().astype(int)

# --- Encodings ---
# Sex: LabelEncoder (maps alphabetically; typically 'female'->0, 'male'->1)
if "Sex" in df_clean.columns:
    # Fill any rare missing values before encoding
    df_clean["Sex"] = df_clean["Sex"].fillna("unknown").astype(str)
    sex_le = LabelEncoder()
    df_clean["Sex_enc"] = sex_le.fit_transform(df_clean["Sex"])

# Embarked: One-hot
if "Embarked" in df_clean.columns:
    # Fill any rare missing values (should be none after imputation)
    df_clean["Embarked"] = df_clean["Embarked"].fillna("S").astype(str)
    df_clean = pd.get_dummies(df_clean, columns=["Embarked"], prefix="Emb", drop_first=False)

print("After cleaning shape:", df_clean.shape)
display(df_clean.head())
```

## 5. Data Visualization & Insights

```python
# Description: Univariate, bivariate, and multivariate visual analysis with matplotlib.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ---------- Univariate Distributions ----------
# Numeric histograms
for col in ["Age", "Fare", "SibSp", "Parch"]:
    if col in df_clean.columns:
        plt.figure()
        plt.hist(df_clean[col].dropna(), bins=30)
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()

# Categorical bar charts
def bar_counts(series, title):
    counts = series.value_counts(dropna=False)
    plt.figure()
    counts.plot(kind="bar")
    plt.title(title)
    plt.xlabel(series.name)
    plt.ylabel("Count")
    plt.show()

for col in ["Pclass", "Sex"]:
    if col in df_clean.columns:
        bar_counts(df_clean[col], f"Counts of {col}")

# Embarked dummies (booleans) -> convert to int for readability
emb_cols = [c for c in ["Emb_C", "Emb_Q", "Emb_S"] if c in df_clean.columns]
if emb_cols:
    plt.figure()
    df_clean[emb_cols].astype(int).sum(axis=0).plot(kind="bar")
    plt.title("Counts of Embarked (one-hot)")
    plt.xlabel("Embarked dummy")
    plt.ylabel("Count")
    plt.show()

# ---------- Bivariate: Survival vs. Key Features ----------
def survival_rate_by(col):
    """Return survival rate (%) grouped by column."""
    grp = df_clean.groupby(col)["Survived"].mean().sort_values(ascending=False) * 100
    return grp

# Survival by Pclass
if "Pclass" in df_clean.columns:
    sr_pclass = survival_rate_by("Pclass")
    plt.figure()
    sr_pclass.plot(kind="bar")
    plt.title("Survival Rate by Pclass (%)")
    plt.xlabel("Pclass")
    plt.ylabel("Survival Rate (%)")
    plt.show()
    print("Survival rate by Pclass (%):\n", sr_pclass.round(1))

# Survival by Sex
if "Sex" in df_clean.columns:
    sr_sex = survival_rate_by("Sex")
    plt.figure()
    sr_sex.plot(kind="bar")
    plt.title("Survival Rate by Sex (%)")
    plt.xlabel("Sex")
    plt.ylabel("Survival Rate (%)")
    plt.show()
    print("\nSurvival rate by Sex (%):\n", sr_sex.round(1))

# Survival by Embarked (reconstruct from dummies if needed)
if set(emb_cols) == set(["Emb_C", "Emb_Q", "Emb_S"]):
    embarked_series = (
        df_clean.assign(
            Embarked_cat=np.select(
                [
                    df_clean["Emb_C"]==True,
                    df_clean["Emb_Q"]==True,
                    df_clean["Emb_S"]==True
                ],
                ["C","Q","S"],
                default="Unknown"
            )
        )["Embarked_cat"]
    )
    sr_emb = df_clean.join(embarked_series.rename("Embarked_cat")).groupby("Embarked_cat")["Survived"].mean()*100
    plt.figure()
    sr_emb.plot(kind="bar")
    plt.title("Survival Rate by Embarked (%)")
    plt.xlabel("Embarked")
    plt.ylabel("Survival Rate (%)")
    plt.show()
    print("\nSurvival rate by Embarked (%):\n", sr_emb.round(1))

# Age distributions by survival (overlaid histograms)
if {"Age","Survived"}.issubset(df_clean.columns):
    plt.figure()
    plt.hist(df_clean.loc[df_clean["Survived"]==0, "Age"].dropna(), bins=30, alpha=0.7)
    plt.hist(df_clean.loc[df_clean["Survived"]==1, "Age"].dropna(), bins=30, alpha=0.7)
    plt.title("Age Distribution by Survival (overlaid)")
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.show()

# Fare by survival (boxplot)
if {"Fare","Survived"}.issubset(df_clean.columns):
    plt.figure()
    data_0 = df_clean.loc[df_clean["Survived"]==0, "Fare"].dropna()
    data_1 = df_clean.loc[df_clean["Survived"]==1, "Fare"].dropna()
    plt.boxplot([data_0, data_1], labels=["Did not survive","Survived"], showfliers=False)
    plt.title("Fare by Survival (Boxplot)")
    plt.ylabel("Fare")
    plt.show()

# Stacked bar: Survival by Pclass & Sex (counts)
if {"Pclass","Sex","Survived"}.issubset(df_clean.columns):
    ctab = pd.crosstab([df_clean["Pclass"], df_clean["Sex"]], df_clean["Survived"])
    ctab.plot(kind="bar", stacked=True)
    plt.title("Survival Counts by Pclass & Sex (Stacked)")
    plt.xlabel("Pclass, Sex")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

# ---------- Multivariate: Correlation Heatmap ----------
numeric_df = df_clean.select_dtypes(include=[np.number])
corr = numeric_df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr, interpolation="nearest")
plt.title("Correlation Heatmap (Numeric Features)")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()

corr.round(2)
```

## Title: Key Findings Summary

```python
# Title: Key Findings Summary
# Description: Concise, data-driven conclusions from this EDA.

findings = [
    "- Higher-class passengers (Pclass=1) show higher survival rates than lower classes.",
    "- Sex is a strong discriminator: females have a substantially higher survival rate than males.",
    "- Embarked location shows differences in survival; distributions may reflect socio-economic or boarding effects.",
    "- Fare correlates with both class and survival; higher fares tend to associate with higher survival.",
    "- Age distributions differ between survivors and non-survivors; very young passengers may have benefited.",
    "- Engineered indicators (e.g., HasCabin) can capture signal from sparse columns."
]
print("=== Key Findings ===")
for f in findings:
    print(f)
```

1 â€” Data Loading

```python
# Description: Load Titanic dataset from a public GitHub source.

import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df2 = pd.read_csv(url)

print("Dataset shape:", df2.shape)
df2.head()
```

## 2 â€” Data Inspection

```python
# Description: Inspect dataset structure, types, and missing values.

print("\n=== Info ===")
df2.info()

print("\n=== Missing values per column ===")
print(df2.isna().sum())

print("\n=== Numeric statistics ===")
display(df2.describe())

print("\n=== Categorical statistics ===")
display(df2.describe(include=['object']))
```

## 3 â€” Data Cleaning

```python
# Description: Handle missing values and prepare dataset for analysis.

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

df2_clean = df2.copy()

# --- Impute Age (median) ---
if "Age" in df2_clean.columns:
    age_imp = SimpleImputer(strategy="median")
    df2_clean["Age"] = age_imp.fit_transform(df2_clean[["Age"]])

# --- Impute Embarked (most frequent) ---
if "Embarked" in df2_clean.columns:
    # Convert to object and replace pandas <NA> with np.nan
    df2_clean["Embarked"] = df2_clean["Embarked"].astype(object).replace({pd.NA: np.nan})

    if df2_clean["Embarked"].isna().all():
        # If all values missing, set default
        df2_clean["Embarked"] = "S"
    else:
        # Fill missing values with the most frequent value (mode)
        df2_clean["Embarked"] = df2_clean["Embarked"].fillna(df2_clean["Embarked"].mode()[0])

    # Ensure consistent string dtype
    df2_clean["Embarked"] = df2_clean["Embarked"].astype(str)

# --- Create HasCabin indicator ---
if "Cabin" in df2_clean.columns:
    df2_clean["HasCabin"] = df2_clean["Cabin"].notna().astype(int)

# --- Encode Sex (binary encoding) ---
if "Sex" in df2_clean.columns:
    sex_le = LabelEncoder()
    df2_clean["Sex_enc"] = sex_le.fit_transform(df2_clean["Sex"])

print("After cleaning shape:", df2_clean.shape)
df2_clean.head()
```

4 â€” Exploratory Analysis

```python
# Description: Perform univariate, bivariate, and multivariate analysis.

import matplotlib.pyplot as plt

# --- Univariate ---
num_cols = ["Age", "Fare", "SibSp", "Parch"]
for col in num_cols:
    plt.figure()
    plt.hist(df2_clean[col], bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# --- Bivariate: Survival vs Pclass ---
sr_pclass = df2_clean.groupby("Pclass")["Survived"].mean() * 100
plt.figure()
sr_pclass.plot(kind="bar")
plt.title("Survival Rate by Pclass (%)")
plt.xlabel("Pclass")
plt.ylabel("Survival Rate (%)")
plt.show()

# --- Bivariate: Survival vs Sex ---
sr_sex = df2_clean.groupby("Sex")["Survived"].mean() * 100
plt.figure()
sr_sex.plot(kind="bar")
plt.title("Survival Rate by Sex (%)")
plt.xlabel("Sex")
plt.ylabel("Survival Rate (%)")
plt.show()

# --- Multivariate: Correlation matrix ---
numeric_df = df2_clean.select_dtypes(include=[np.number])
corr = numeric_df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr, interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

5 â€” Insight Generation

```python
# Description: Summarize main findings from the analysis.

insights = [
    "Higher-class passengers (Pclass=1) had a much higher survival rate than lower classes.",
    "Females were significantly more likely to survive than males.",
    "Fare is positively correlated with survival â€” higher fares generally mean better survival chances.",
    "HasCabin indicator may also be linked to higher survival."
]
print("=== Key Insights ===")
for i in insights:
    print("-", i)
```

## Exercise 3 â€” Iris Dataset: Exploratory Data Analysis

**Goal:** Explore the Iris dataset, visualize distributions and relationships, and extract key insights that differentiate species.

**Steps**

## 1 â€” Data Loading

```python
# Description: Load the Iris dataset from scikit-learn and convert to a pandas DataFrame.

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df_iris = iris.frame.copy()

# Rename columns to match the exercise specification
rename_map = {
    "sepal length (cm)": "SepalLengthCm",
    "sepal width (cm)": "SepalWidthCm",
    "petal length (cm)": "PetalLengthCm",
    "petal width (cm)": "PetalWidthCm",
    "target": "Species"
}
df_iris = df_iris.rename(columns=rename_map)

# Map numeric species to string labels
species_map = {i: name for i, name in enumerate(iris.target_names)}
df_iris["Species"] = df_iris["Species"].map(species_map)

print("Dataset shape:", df_iris.shape)
df_iris.head()
```

## 2 â€” Data Inspection


```python
# Description: Inspect structure, types, statistics, and class distribution.

print("=== .info() ===")
df_iris.info()

print("\n=== Descriptive statistics (numeric) ===")
display(df_iris.describe())

print("\n=== Unique species and counts ===")
print(df_iris["Species"].value_counts())

print("\n=== Missing values per column ===")
print(df_iris.isna().sum())
```

## 3 â€” Data Visualization (Histograms & Scatter)

```python
# Description: Plot univariate histograms and pairwise scatterplots.

import matplotlib.pyplot as plt
import itertools

# Histograms for each feature
for col in ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]:
    plt.figure()
    plt.hist(df_iris[col], bins=20)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# Simple pairwise scatter plots (no seaborn)
features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
pairs = list(itertools.combinations(features, 2))

# For each pair, create a scatter colored by species using separate series
for x_col, y_col in pairs:
    plt.figure()
    for sp in df_iris["Species"].unique():
        subset = df_iris[df_iris["Species"] == sp]
        plt.scatter(subset[x_col], subset[y_col], label=sp)
    plt.title(f"{y_col} vs {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.show()
```

## 4 â€” Exploratory Analysis (Uni/Bi/Multivariate)

```python
# Description: Univariate per-species stats, bivariate group comparisons, and multivariate correlations.

import numpy as np
import pandas as pd

# --- Univariate per species: means and stds ---
stats = (
    df_iris.groupby("Species")[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
    .agg(["mean","std","min","max"])
    .round(2)
)
print("=== Per-Species Summary (mean/std/min/max) ===")
display(stats)

# --- Bivariate: simple differences in means between species (example) ---
mean_by_species = df_iris.groupby("Species").mean(numeric_only=True).round(2)
print("\n=== Mean by Species ===")
display(mean_by_species)

# --- Multivariate: correlation matrix (numeric only) ---
corr = df_iris[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].corr(numeric_only=True)

import matplotlib.pyplot as plt
plt.figure()
plt.imshow(corr, interpolation="nearest")
plt.title("Correlation Heatmap (Iris)")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()

corr.round(2)
```

## 5 â€” Insight Generation

```python
# Description: Summarize the key patterns that differentiate species.

insights = [
    "- Petal measurements (length and width) show the strongest separation among species.",
    "- Setosa typically has much smaller petals than Versicolour and Virginica.",
    "- Virginica tends to have the largest petal measurements overall.",
    "- Sepal measurements separate species less cleanly compared to petals, but still provide signal.",
    "- Strong positive correlation exists between PetalLengthCm and PetalWidthCm."
]
print("=== Key Insights ===")
for i in insights:
    print(i)
```

 ## 6 â€” Advanced Visualization (Boxplots & Grouped Means)

```python
# Description: Compare distributions across species using boxplots and grouped bar charts.

import matplotlib.pyplot as plt
import numpy as np

features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]

# Boxplots per feature across species
for col in features:
    plt.figure()
    data_by_species = [df_iris[df_iris["Species"]==sp][col] for sp in df_iris["Species"].unique()]
    plt.boxplot(data_by_species, labels=df_iris["Species"].unique(), showfliers=False)
    plt.title(f"{col} by Species (Boxplot)")
    plt.xlabel("Species")
    plt.ylabel(col)
    plt.show()

# Grouped means per species (bar chart)
means = df_iris.groupby("Species")[features].mean().T  # shape: features x species
plt.figure()
means.plot(kind="bar")
plt.title("Mean Feature Values by Species")
plt.xlabel("Feature")
plt.ylabel("Mean")
plt.tight_layout()
plt.show()
```

# Conclusion

The exploratory data analysis of the Iris dataset reveals clear and consistent patterns that distinguish the three species â€” Setosa, Versicolour, and Virginica â€” based on sepal and petal measurements.

Petal dimensions, in particular, demonstrate the strongest discriminatory power. Setosa consistently shows smaller petal lengths and widths compared to the other two species, while Virginica generally has the largest petal measurements. Versicolour occupies an intermediate position but remains well separated from Setosa due to significantly larger petal dimensions. The strong positive correlation between petal length and petal width reinforces their importance as key features for species classification.

Sepal measurements, while less distinct, still provide useful separation in combination with petal features. Sepal width shows moderate variation between species, and sepal length can contribute to fine-tuning classification boundaries.

From a multivariate perspective, the dataset presents a highly structured pattern, with minimal overlap between species when visualized in two-dimensional feature space (especially for petal measurements). This structure supports the suitability of the Iris dataset for demonstrating supervised machine learning techniques, as well as for teaching and practicing fundamental data analysis skills.

Overall, the Iris dataset remains an exemplary case study for understanding the relationship between data features and class labels, and for illustrating the full EDA process from univariate exploration to multivariate pattern identification.
