### Exercise 1: Understanding Data Visualization

#### Task 1: Why is data visualization important in data analysis?
Answer:
Data visualization is essential in data analysis because it helps transform complex data into clear and understandable visuals. It allows analysts to:
Quickly identify trends, patterns, and outliers.
Communicate findings effectively to stakeholders.
Make data-driven decisions based on visual evidence.
Spot potential errors or anomalies that may not be visible in raw data.

In short, visualizations turn raw numbers into meaningful insights.

#### Task 2: What is the purpose of a line graph in data visualization?
Answer:
A line graph is used to show how a variable changes over time or across continuous values. It is ideal for tracking trends, such as stock prices, sales over months, or temperature changes. Line graphs help:
Visualize the direction and rate of change.
Compare multiple trends on the same chart.
Detect growth patterns, peaks, or declines over time.
It’s especially helpful when analyzing time-series data.

### Exercise 2: Creating a Line Plot for Temperature Variation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Temperature data for each day of the week
temperatures = [72, 74, 76, 80, 82, 78, 75]
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# Create the line plot
plt.plot(days, temperatures, marker='o', linestyle='-', color='orange')

# Add labels and title
plt.xlabel("Day")
plt.ylabel("Temperature (°F)")
plt.title("Temperature Variation Over a Week")

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("Student_Mental_health.csv")
```

```python
print(df.head())
```

```python
print(df.describe(include='all'))
```

```python
print(df.info())
```

```python
# Définir une palette personnalisée : rose pastel pour femmes, bleu pastel pour hommes
custom_palette = {"Female": "lightpink", "Male": "lightblue"}

plt.figure(figsize=(8, 5))

# Tracer un barplot avec "Do you have Anxiety?" sur l'axe x, et hue = genre
sns.countplot(data=df, x="Do you have Anxiety?", hue="Choose your gender", palette=custom_palette)

# Personnaliser
plt.title("Anxiety Responses by Gender")
plt.xlabel("Do you have Anxiety?")
plt.ylabel("Number of Students")
plt.legend(title="Gender")
plt.show()
```

```python
# Calculer le nombre total de réponses par genre
total_by_gender = df['Choose your gender'].value_counts()

# Calculer les réponses croisées "Anxiety" x "Gender"
cross_counts = df.groupby(['Do you have Anxiety?', 'Choose your gender']).size().reset_index(name='Count')

# Ajouter la colonne de proportion en %
cross_counts['Proportion'] = cross_counts.apply(
    lambda row: 100 * row['Count'] / total_by_gender[row['Choose your gender']], axis=1
)

# Palette personnalisée en pastel
palette = {"Female": "lightpink", "Male": "lightblue"}

# Affichage du graphique
plt.figure(figsize=(8, 5))
sns.barplot(
    data=cross_counts,
    x="Do you have Anxiety?",
    y="Proportion",
    hue="Choose your gender",
    palette=palette
)

plt.title("Proportion of Anxiety Responses by Gender (%)")
plt.ylabel("Percentage")
plt.xlabel("Do you have Anxiety?")
plt.ylim(0, 100)
plt.legend(title="Gender")
plt.show()
```

### Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Load data
df = pd.read_csv('Student_Mental_health.csv')
```

```python
# Clean and prepare
df = df.dropna(subset=['Age', 'Do you have Panic attack?'])
df['Age'] = df['Age'].astype(int)
df['Panic_numeric'] = df['Do you have Panic attack?'].map({'Yes': 'Yes', 'No': 'No'})
```

```python
ages_present = sorted(df['Age'].unique())

counts = df.groupby(['Age', 'Panic_numeric']).size().reset_index(name='count')
counts = counts[counts['Age'].isin(ages_present)]
total_per_age = counts.groupby('Age')['count'].transform('sum')
counts['percent'] = counts['count'] / total_per_age * 100
```

```python
plt.figure(figsize=(12,6))

palette = {'No': '#90ee90', 'Yes': '#ff7f7f'}

ax = sns.barplot(data=counts, x='Age', y='percent', hue='Panic_numeric',
                 order=ages_present, palette=palette)

# Calculate weighted average
total_counts = df.groupby('Age').size().reindex(ages_present)
yes_data = counts[counts['Panic_numeric'] == 'Yes'].set_index('Age').reindex(ages_present).fillna(0)
yes_percent = yes_data['percent'].values
weighted_avg = np.sum(yes_percent * total_counts.values) / total_counts.sum()

# Plot pink dashed weighted average line without label
plt.axhline(weighted_avg, color='#ff69b4', linestyle='--', linewidth=1)

plt.xlabel("Age")
plt.ylabel("Percentage (%)")
plt.title("Percentage of Students With and Without Panic Attacks by Age")

# Get handles and labels from barplot
handles, labels = ax.get_legend_handles_labels()

# Create custom handle for weighted average line
weighted_avg_handle = Line2D([], [], color='#ff69b4', linestyle='--', linewidth=1)

# Add weighted average handle and label to legend
handles.append(weighted_avg_handle)
labels.append(f'Weighted Average {weighted_avg:.2f}%')

plt.legend(handles=handles, labels=labels, title='Panic Attacks')

ax.set_xticks(np.arange(len(ages_present)))
ax.set_xticklabels([str(age) for age in ages_present])

plt.show()
```

The bar chart displays the percentage of students reporting panic attacks ("Yes") versus those who do not ("No") across different ages. The green bars represent students without panic attacks, while the light red bars represent those with panic attacks.

A pink dashed horizontal line indicates the weighted average percentage of students experiencing panic attacks, accounting for the number of students at each age group.
