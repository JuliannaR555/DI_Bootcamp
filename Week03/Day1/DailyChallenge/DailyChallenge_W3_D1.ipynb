#### 1. Generate Synthetic Temperature Data

```python
# 1. Fix the seed for reproducibility
np.random.seed(0)

# 2. Generate the temperature array: 10 cities × 12 months
temperatures = np.random.uniform(low=-5, high=35, size=(10, 12))

# 3. Create the DataFrame
cities = ['Paris', 'Cairo', 'Tokyo', 'New York', 'Sydney', 'Moscow', 'Lima', 'Cape Town', 'Beijing', 'Reykjavik']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

df_temp = pd.DataFrame(data=temperatures, index=cities, columns=months)
```

#### 2. Explore the Dataset

```python
# Basic statistics
print("\nDescriptive statistics:")
df_temp.describe()
```

#### 3. Analyze Annual Averages

```python
# 1. Calculate the annual average temperature for each city
df_temp['Annual Avg'] = df_temp.mean(axis=1)
```

```python
# 2. Identify the hottest city (max average)
hottest_city = df_temp['Annual Avg'].idxmax()
hottest_temp = df_temp['Annual Avg'].max()
```

```python
# 3. Identify the coldest city (min average)
coldest_city = df_temp['Annual Avg'].idxmin()
coldest_temp = df_temp['Annual Avg'].min()
```

```python
# 4. Display the results
print("Annual average temperature by city:")
print(df_temp['Annual Avg'])
```

```python
print(f"\nHottest city of the year: {hottest_city} ({hottest_temp:.2f} °C)")
print(f"Coldest city of the year: {coldest_city} ({coldest_temp:.2f} °C)")
```

#### 4. Visualize Monthly Trends

```python
# Exclude the "Annual Avg" column
months_only = df_temp.columns[:-1]

# Create line plot for each city
plt.figure(figsize=(14, 8))

for city in df_temp.index:
    plt.plot(months_only, df_temp.loc[city, months_only], label=city, linewidth=2)
    plt.savefig("temperature_trends.png", dpi=300, bbox_inches='tight')

plt.title("Monthly Temperature Trends for Each City")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Summary Report (Updated from Visual Interpretation)

This project simulated and analyzed average monthly temperatures for 10 cities over a full year, using NumPy for data generation and Pandas/Matplotlib for analysis and visualization.

### Temperature Insights by City

From the line chart above, we can observe several key temperature patterns:

- **Reykjavik**, despite being expected to be cold, shows relatively **stable temperatures** throughout the year in this synthetic data — with warmer peaks in summer months. This makes it the hottest city *in this dataset* (due to random generation).

- **Cape Town** shows a nearly **inverted pattern** compared to cities in the Northern Hemisphere, due to its location in the Southern Hemisphere (i.e., warmer around December–February).
