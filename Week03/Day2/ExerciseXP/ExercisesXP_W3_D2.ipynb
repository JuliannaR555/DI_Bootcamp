#### Exercise 1 : Matrix Operations

```python
import numpy as np

# Step 1: Define a 3x3 matrix
A = np.array([[3, 0, 2],
              [2, 0, -2],
              [0, 1, 1]])

print("Matrix A:\n", A)

# Step 2: Calculate the determinant
det_A = np.linalg.det(A)
print("Determinant of A:", det_A)

# Step 3: Check if the matrix is invertible
if det_A != 0:
    # Step 4: Compute the inverse
    A_inv = np.linalg.inv(A)
    print("Inverse of A:\n", A_inv)
else:
    print("The matrix is not invertible.")
```

#### Exercise 6 : Basic Hypothesis Testing

```python
import numpy as np

# Step 1: Generate synthetic productivity data
np.random.seed(42)
productivity_before = np.random.normal(loc=50, scale=10, size=30)
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Step 2: Compute means
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print("Mean before training:", mean_before)
print("Mean after training:", mean_after)

# Step 3: Compute the mean difference
mean_diff = np.mean(productivity_after - productivity_before)
print("Average improvement:", mean_diff)

# Step 4 (optional): Compute a simple effect size
std_diff = np.std(productivity_after - productivity_before)
print("Standard deviation of improvement:", std_diff)

# Step 5: Check basic condition
if mean_diff > 0:
    print(" The program likely had a positive effect.")
else:
    print(" The program did not improve productivity.")
```

```python
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Test t appari : compare les moyennes des scores lis (mme groupe avant/aprs)
t_stat, p_value = ttest_rel(productivity_after, productivity_before)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interprtation
alpha = 0.05
if p_value < alpha:
    print(" We reject H0  The training program significantly improved productivity.")
else:
    print(" We fail to reject H0  No significant effect detected.")
```

```python
# Botes  moustaches (boxplots) pour visualiser la diffrence
plt.boxplot([productivity_before, productivity_after], tick_labels=['Before', 'After'])
plt.title("Employee Productivity Before vs After Training")
plt.ylabel("Productivity Score")
plt.grid(True)
plt.show()
```

The training program had a statistically significant positive effect on employee productivity.
The average productivity increased by +4.64 points, and the p-value (5.47e-10) confirms that this improvement is unlikely due to chance.
We reject the null hypothesis.
