## 1. What is data analysis?

## Data analysis is the process of examining, organizing, cleaning, transforming, and interpreting raw data in order to uncover meaningful patterns, trends, or relationships that can guide decision-making. It bridges the gap between data collection and action. Whether we are trying to understand customer behavior, improve healthcare outcomes, or optimize a supply chain, data analysis provides the insight needed to move forward intelligently.

## There are different types of data analysis, including descriptive (what happened?), diagnostic (why did it happen?), predictive (what might happen?), and prescriptive (what should we do?). Data analysis is supported by a variety of tools and methods, from simple spreadsheets to complex machine learning algorithms.

## In the modern digital era, organizations across every industry collect vast amounts of data—from web traffic and sales figures to biometric sensors and GPS trackers. Without analysis, this data is useless. With it, we can predict disease outbreaks, reduce costs, detect fraud, personalize services, and even design smarter cities.

## Modern trends in data analysis include the increasing use of artificial intelligence (AI) and automation, real-time analytics, and the ethical use of data. Technologies like ChatGPT, recommendation engines, and fraud detection systems all rely heavily on data analysis techniques. As the amount of data grows, the importance of data analysis continues to increase.

## Sleep Dataset: Identifying Data Types

## We will classify each column as **quantitative** or **qualitative** based on the kind of data it contains.
## Quantitative: numerical values that can be measured
## Qualitative: categorical or descriptive values

```python
def classify_column(col):
    """
    Classify a column as 'Quantitative' if it contains numeric data,
    or 'Qualitative' if it contains text or categorical data.

    Parameters:
    col (pd.Series): A column from a DataFrame

    Returns:
    str: 'Quantitative' or 'Qualitative'
    """
    if pd.api.types.is_numeric_dtype(col):
        return "Quantitative"
    else:
        return "Qualitative"
```

```python
for column in sleep_df.columns:
    col_type = classify_column(sleep_df[column])
    print(f"'{column}' ➜ {col_type}")
```

### Column Classification: Sleep Dataset

## `index`: Quantitative: It is a numeric index used to identify each row.
## `Year`: Quantitative: Represents the calendar year (numeric).
## `Period`: Qualitative: Indicates a time category (e.g., monthly, quarterly).
## `Avg hrs per day sleeping`: Quantitative: Numeric value showing average sleep duration.
## `Standard Error`: Quantitative: A statistical measure (numeric).
## `Type of Days`: Qualitative: Describes day type (e.g., weekday, weekend).
## `Age Group`: Qualitative: Categorical group (e.g., 15-24, 25-34).
## `Activity`: Qualitative: Name of the recorded activity (text).
## `Sex`: Qualitative: Gender of participants (e.g., Male, Female).

## Exercise 5: Basic Data Analysis with Google Colab

### Step 1: Calculate Basic Statistics


```python
mean_petal = iris_df['PetalLengthCm'].mean()
median_petal = iris_df['PetalLengthCm'].median()
mode_petal = iris_df['PetalLengthCm'].mode()

print("Mean (Petal Length):", mean_petal)
print("Median (Petal Length):", median_petal)
print("Mode (Petal Length):", mode_petal.tolist())
```

## Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=iris_df, x='PetalLengthCm', kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.show()
```
