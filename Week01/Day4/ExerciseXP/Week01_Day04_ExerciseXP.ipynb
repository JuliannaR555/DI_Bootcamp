Exercise 1: Pets

```python
# 1 : Définition de la classe Pets
class Pets():
    def __init__(self, animals):
        # On stocke la liste des animaux dans un attribut
        self.animals = animals

    def walk(self):
        # Pour chaque animal, on appelle sa méthode walk()
        for animal in self.animals:
            print(animal.walk())

# Classe de base pour les chats
class Cat():
    is_lazy = True  # Tous les chats sont paresseux

    def __init__(self, name, age):
        # Chaque chat a un nom et un âge
        self.name = name
        self.age = age

    def walk(self):
        # Retourne une phrase disant que le chat marche
        return f'{self.name} is just walking around'


# Classe Bengal qui hérite de Cat
class Bengal(Cat):
    def sing(self, sounds):
        return f'{sounds}'

# Classe Chartreux qui hérite de Cat
class Chartreux(Cat):
    def sing(self, sounds):
        return f'{sounds}'

# 1 : Création de la classe Siamese qui hérite de Cat
class Siamese(Cat):
    def sing(self, sounds):
        return f'{self.name} says: {sounds}'
    
    # Méthode supplémentaire pour faire ronronner le chat
    def purr(self):
        return f'{self.name} is purring softly'


# 2 : Création des instances de chats avec nom et âge
bengal_cat = Bengal("Leo", 3)
chartreux_cat = Chartreux("Misty", 5)
siamese_cat = Siamese("Luna", 2)

# Liste de tous les chats
all_cats = [bengal_cat, chartreux_cat, siamese_cat]

# 3 : Création d'une instance de Pets avec les chats
sara_pets = Pets(all_cats)

# 4 : Promener les chats (appel de la méthode walk)
sara_pets.walk()
```

Exercise 2: Dogs

```python
# 1 : Création de la classe Dog
class Dog:
    def __init__(self, name, age, weight):
        # Initialisation des attributs nom, âge et poids
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        # Retourne une phrase quand le chien aboie
        return f"{self.name} is barking"

    def run_speed(self):
        # Calcule la vitesse de course du chien
        return (self.weight / self.age) * 10

    def fight(self, other_dog):
        # Calcule la puissance de chaque chien : vitesse * poids
        self_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight

        # Compare les puissances pour déterminer le gagnant
        if self_power > other_power:
            return f"{self.name} won the fight against {other_dog.name}"
        elif self_power < other_power:
            return f"{other_dog.name} won the fight against {self.name}"
        else:
            return f"{self.name} and {other_dog.name} are equally strong!"


# 2 : Création de 3 instances de Dog
dog1 = Dog("Rex", 5, 20)
dog2 = Dog("Bella", 3, 15)
dog3 = Dog("Thor", 4, 25)

# 3 : Test des méthodes

# Le chien 1 aboie
print(dog1.bark())  # Rex is barking

# Vitesse de course du chien 2
print(f"{dog2.name} runs at speed: {dog2.run_speed()}")

# Combat entre le chien 1 et le chien 3
print(dog1.fight(dog3))

# Combat entre le chien 2 et le chien 3
print(dog2.fight(dog3))
```

Exercise 3: Dogs Domesticated

```python
# 1 : Importer le module random pour choisir un tour au hasard
import random

# On suppose que la classe Dog est dans le même fichier ou a été copiée ici :
class Dog:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        return f"{self.name} is barking"

    def run_speed(self):
        return (self.weight / self.age) * 10

    def fight(self, other_dog):
        self_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight

        if self_power > other_power:
            return f"{self.name} won the fight against {other_dog.name}"
        elif self_power < other_power:
            return f"{other_dog.name} won the fight against {self.name}"
        else:
            return f"{self.name} and {other_dog.name} are equally strong!"

# 2 : Création de la classe PetDog qui hérite de Dog
class PetDog(Dog):
    def __init__(self, name, age, weight):
        # On appelle le constructeur de la classe parente Dog
        super().__init__(name, age, weight)
        # Nouvel attribut : est-ce que le chien est entraîné ?
        self.trained = False

    def train(self):
        # Le chien aboie et devient entraîné
        print(self.bark())
        self.trained = True

    def play(self, *args):
        # Affiche les noms des chiens qui jouent ensemble
        dog_names = ', '.join(args)
        print(f"{self.name}, {dog_names} all play together")

    def do_a_trick(self):
        # Si le chien est entraîné, il fait un tour choisi au hasard
        if self.trained:
            tricks = [
                "does a barrel roll",
                "stands on his back legs",
                "shakes your hand",
                "plays dead"
            ]
            print(f"{self.name} {random.choice(tricks)}")
        else:
            print(f"{self.name} doesn't know any tricks yet.")


# 3 : Tester les méthodes

# Création d’un chien domestique
my_dog = PetDog("Fido", 2, 10)

# Entraînement
my_dog.train()

# Il joue avec d'autres chiens
my_dog.play("Buddy", "Max")

# Il fait un tour
my_dog.do_a_trick()
```

Exercise 4: Family and Person Classes

```python
# 1 : Définir la classe Personne
class Person:
    def __init__(self, first_name, age):
        self.first_name = first_name
        self.age = age
        self.last_name = ""  # Attribut initialisé vide

    def is_18(self):
        # Retourne True si la personne a 18 ans ou plus
        return self.age >= 18


# 2 : Définir la classe Famille
class Family:
    def __init__(self, last_name):
        self.last_name = last_name
        self.members = []  # Liste de personnes dans la famille

    def born(self, first_name, age):
        # Crée une nouvelle personne
        new_member = Person(first_name, age)
        # Attribue le nom de famille à cette personne
        new_member.last_name = self.last_name
        # Ajoute cette personne à la famille
        self.members.append(new_member)

    def check_majority(self, first_name):
        # Recherche la personne par prénom
        for member in self.members:
            if member.first_name == first_name:
                if member.is_18():
                    print(f"You are over 18, your parents Jane and John accept that you will go out with your friends")
                else:
                    print(f"Sorry, you are not allowed to go out with your friends.")
                return
        # Si aucune personne ne correspond au prénom donné
        print(f"No family member named {first_name} was found.")

    def family_presentation(self):
        # Affiche le nom de famille
        print(f"Family name: {self.last_name}")
        print("Family members:")
        # Affiche chaque membre avec son prénom et son âge
        for member in self.members:
            print(f"{member.first_name}, {member.age} years old")
```

```python
# Création d'une famille
romanova_family = Family("Romanova")

# Naissance de membres
romanova_family.born("Chloe", 8)
romanova_family.born("Julianna", 43)
romanova_family.born("Elena", 17)

# Vérification de la majorité
romanova_family.check_majority("Chloe")
romanova_family.check_majority("Elena")
romanova_family.check_majority("Julianna")

# Présentation de la famille
romanova_family.family_presentation()
```
