```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#Load and display one image
image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)
```

```python
#Display the original image using matplotlib
plt.imshow(original_image)
plt.axis('off')        #Hide axes
plt.title("Original Image")
plt.show()             #Actually show the image
```

```python
from scipy.ndimage import rotate
import numpy as np

#Rotate the image by 30 degrees (as requested in the instruction to complete the function)
def rotate_image_30_degrees(image):
    return rotate(np.array(image), 30, reshape=False, mode='nearest')

rotated_30 = rotate_image_30_degrees(original_image)

plt.imshow(rotated_30)
plt.axis('off')
plt.title("Rotated 30°")
plt.show()
```

######Rotate the image by 30 degrees using scipy.ndimage.rotate

```python
#Rotate the image by 90 degrees
def rotate_image_90_degrees(image):
    return rotate(np.array(image), 90, reshape=False, mode='nearest')

rotated_90 = rotate_image_90_degrees(original_image)

plt.imshow(rotated_90)
plt.axis('off')
plt.title("Rotated 90°")
plt.show()
```

##### Rotate the image by 90 degrees using a separate function

```python
from PIL import ImageOps

#Flip horizontally (mirror)
flipped_horizontal = ImageOps.mirror(original_image)

plt.imshow(flipped_horizontal)
plt.axis('off')
plt.title("Flipped Horizontally")
plt.show()
```

##### Flip the image horizontally using ImageOps.mirror

```python
#Flip vertically
flipped_vertical = ImageOps.flip(original_image)

plt.imshow(flipped_vertical)
plt.axis('off')
plt.title("Flipped Vertically")
plt.show()
```

#####Flip the image vertically using ImageOps.flip

```python
#Zoom in on the image by 1.2x using resize()
width, height = original_image.size
zoomed_image = original_image.resize((int(width * 1.2), int(height * 1.2)))

plt.imshow(zoomed_image)
plt.axis('off')
plt.title("Zoomed ×1.2")
plt.show()
```
